自己寫function 包括測試分集

```python

import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('cancer.txt')

def custom_train_test_split(data, test_size=0.2):

    num_samples = len(data)
    num_test_samples = int(test_size * num_samples)

    X_train = data[num_test_samples:, :-1]  # 特徵欄位
    y_train = data[num_test_samples:, -1]   # 分類標籤

    X_test = data[:num_test_samples, :-1]
    y_test = data[:num_test_samples, -1]

    return X_train, X_test, y_train, y_test

# 使用自訂的函式切分資料集
x_train, x_test, y_train, y_test = custom_train_test_split(data, test_size=0.2)



# 假設 x_train, x_test, y_train, y_test 已經準備好

# 初始化權重和偏差
input_neurons = x_train.shape[1]
hidden_neurons_1 = 15
hidden_neurons_2 = 10
output_neurons = 5

np.random.seed(42)
weights_input_hidden1 = np.random.randn(input_neurons, hidden_neurons_1) * 0.5
bias_hidden1 = np.zeros((1, hidden_neurons_1))
weights_hidden1_hidden2 = np.random.randn(hidden_neurons_1, hidden_neurons_2) * 0.01
bias_hidden2 = np.zeros((1, hidden_neurons_2))
weights_hidden2_output = np.random.randn(hidden_neurons_2, output_neurons) * 0.01
bias_output = np.zeros((1, output_neurons))

# 設定訓練迭代次數及學習率
epochs = 100000
learning_rate = 0.01

# 儲存迭代的 training accuracy 和 test accuracy
train_accuracies = []
test_accuracies = []

# 訓練模型
for epoch in range(epochs):
    # Forward Propagation
    hidden_layer_input1 = np.dot(x_train, weights_input_hidden1) + bias_hidden1
    hidden_layer_output1 = 1 / (1 + np.exp(-hidden_layer_input1))

    hidden_layer_input2 = np.dot(hidden_layer_output1, weights_hidden1_hidden2) + bias_hidden2
    hidden_layer_output2 = 1 / (1 + np.exp(-hidden_layer_input2))

    output_layer_input = np.dot(hidden_layer_output2, weights_hidden2_output) + bias_output
    predicted_output = 1 / (1 + np.exp(-output_layer_input))

    # Backpropagation
    error = y_train.reshape(-1, 1) - predicted_output
    d_predicted_output = error * predicted_output * (1 - predicted_output)

    error_hidden2 = d_predicted_output.dot(weights_hidden2_output.T)
    d_hidden2 = error_hidden2 * hidden_layer_output2 * (1 - hidden_layer_output2)

    error_hidden1 = d_hidden2.dot(weights_hidden1_hidden2.T)
    d_hidden1 = error_hidden1 * hidden_layer_output1 * (1 - hidden_layer_output1)

    # Updating Weights and Biases
    weights_hidden2_output += hidden_layer_output2.T.dot(d_predicted_output) * learning_rate
    bias_output += np.sum(d_predicted_output, axis=0, keepdims=True) * learning_rate
    weights_hidden1_hidden2 += hidden_layer_output1.T.dot(d_hidden2) * learning_rate
    bias_hidden2 += np.sum(d_hidden2, axis=0, keepdims=True) * learning_rate
    weights_input_hidden1 += x_train.T.dot(d_hidden1) * learning_rate
    bias_hidden1 += np.sum(d_hidden1, axis=0, keepdims=True) * learning_rate

    # 每 10000 次迭代計算一次準確度
    if epoch % 10000 == 0:
        # 計算訓練集的準確度
        predicted_output_train = 1 / (1 + np.exp(-output_layer_input))
        predicted_output_train_binary = np.round(predicted_output_train)
        train_accuracy = np.mean(predicted_output_train_binary == y_train.reshape(-1, 1))
        train_accuracies.append(train_accuracy)

        # 在測試集上進行預測
        hidden_layer_input1_test = np.dot(x_test, weights_input_hidden1) + bias_hidden1
        hidden_layer_output1_test = 1 / (1 + np.exp(-hidden_layer_input1_test))
        hidden_layer_input2_test = np.dot(hidden_layer_output1_test, weights_hidden1_hidden2) + bias_hidden2
        hidden_layer_output2_test = 1 / (1 + np.exp(-hidden_layer_input2_test))
        output_layer_input_test = np.dot(hidden_layer_output2_test, weights_hidden2_output) + bias_output
        predicted_output_test = 1 / (1 + np.exp(-output_layer_input_test))
        predicted_output_test_binary = np.round(predicted_output_test)
        test_accuracy = np.mean(predicted_output_test_binary == y_test.reshape(-1, 1))
        test_accuracies.append(test_accuracy)

        print(f"Epoch {epoch}: Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
# 繪製訓練曲線
plt.plot(train_accuracies, label='Training Accuracy')
plt.title('Training Curve')
plt.xlabel('Epochs (x10,000)')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
